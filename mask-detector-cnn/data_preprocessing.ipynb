```python
import cv2, os

data_path = 'dataset' #path where the dataset is stored
categories = os.listdir(data_path) # getting the names of the categories
labels = [i for i in range(len(categories))] # labels: 0, 1

label_dict = dict(zip(categories, labels)) # making a dictonary

print(label_dict)
print(categories)
print(labels)
```

```python
data = []
target = []
size = 100
for category in categories:
    folder_path = os.path.join(data_path, category) # getting the path
    img_names = os.listdir(folder_path) #getting the names of the images
    for img_name in img_names:
        img_path = os.path.join(folder_path, img_name) #making the image path
        img = cv2.imread(img_path) #reading the image

        try:
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #converting the image into gray scale
            resized = cv2.resize(gray, (size, size)) # resizing the image
            data.append(resized) #appending the image into data list
            target.append(label_dict[category])# appending the image into target list
        except Exception as e:
            print("Exception: ", e) # in case of any exception is raised, the exceptionwill be printed and the loop will move on to the next program
```

```python
import numpy as np

data = np.array(data) / 255.0 #normalizing the array
data = np.reshape(data, (data.shape[0], size, size, 1)) # reshaping the value
target = np.array(target) # converting the target into a numpy array

from keras.utils import np_utils

new_target = np_utils.to_categorical(target) # converting into categorical representation since the outer most layer has two neurons for categorising the image
```

```python
np.save('data', data) #saving the data
np.save('target', new_target) # saving the images
```

```python
import matplotlib.pyplot as plt
```

```python
mask_folder = os.path.join(data_path, categories[0])
no_mask_folder = os.path.join(data_path, categories[1])
```

```python
mask_li = []
no_mask_li = []
```

```python
mask_images = os.listdir(mask_folder)
for i in range(0, 8):
    temp_path = os.path.join(mask_folder, mask_images[i])
    temp = cv2.imread(temp_path)
    img = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
    mask_li.append(img)
```

```python
no_mask_images = os.listdir(no_mask_folder)
for i in range(0, 8):
    temp_path = os.path.join(no_mask_folder, no_mask_images[i])
    temp = cv2.imread(temp_path)
    img = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
    no_mask_li.append(img)
```

```python
try:
    fig, ax = plt.subplots(2, 7, figsize=(19, 4),
                          subplot_kw={'xticks':[], 'yticks':[]},
                          gridspec_kw=dict(hspace=0.1, wspace=0.1))

    
    for i in range(0,7):
            ax[0, i].imshow(mask_li[i])
            ax[1, i].imshow(no_mask_li[i])
            
    ax[0, 0].set_ylabel('mask\nimages')
    ax[1, 0].set_ylabel('no mask\nimages')
except Exception as e:
    print(e)
```
